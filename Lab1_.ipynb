```python
# =========================================
# 1. Import Required Libraries
# =========================================
import numpy as np
import pandas as pd

# =========================================
# 2. Load Dataset
# =========================================
df = pd.read_csv("/content/parkinsons.data")

print("First 5 Rows of Dataset:")
print(df.head())

# =========================================
# 3. Python Programming Constructs
# =========================================

# List of first 5 values of a column
values = df["MDVP:Fo(Hz)"].head().tolist()

# Loop example
total = 0
for v in values:
    total += v

print("\nSum using loop:", total)

# Condition example
if total > 500:
    print("Sum is greater than 500")
else:
    print("Sum is less than or equal to 500")

# Function example
def square(x):
    return x * x

print("Square of 10:", square(10))


# =========================================
# 4. NumPy Operations
# =========================================

# Convert column to NumPy array
arr = np.array(df["MDVP:Fo(Hz)"])

print("\nNumPy Array Sample:", arr[:5])

print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Max Value:", np.max(arr))
print("Min Value:", np.min(arr))


# =========================================
# 5. Pandas Data Analysis
# =========================================

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

print("\nClass Distribution (Target Variable):")
print(df["status"].value_counts())


# =========================================
# 6. Data Selection & Filtering
# =========================================

# Select specific column
print("\nMDVP:Fo(Hz) Column:")
print(df["MDVP:Fo(Hz)"].head())

# Filter rows where status = 1
print("\nPatients with Parkinson’s (status=1):")
print(df[df["status"] == 1].head())


# =========================================
# 7. Basic Data Cleaning Example
# =========================================

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Drop 'name' column (not needed for analysis)
df.drop("name", axis=1, inplace=True)

print("\nDataset after dropping 'name' column:")
print(df.head())
```
