```python
# =========================================
# 1. Import Libraries
# =========================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity

# =========================================
# 2. Create User-Item Dataset
# =========================================
data = {
    "Movie1": [5, 4, 0, 0],
    "Movie2": [4, 0, 0, 2],
    "Movie3": [0, 0, 5, 4],
    "Movie4": [0, 3, 4, 0]
}

df = pd.DataFrame(data, index=["User1", "User2", "User3", "User4"])

print("User-Movie Rating Matrix:\n")
print(df)

# =========================================
# 3. Compute Similarity Matrix
# =========================================
similarity = cosine_similarity(df)

similarity_df = pd.DataFrame(
    similarity,
    index=df.index,
    columns=df.index
)

print("\nUser Similarity Matrix:\n")
print(similarity_df)

# =========================================
# 4. Visualization (Heatmap)
# =========================================
plt.figure(figsize=(6,5))

sns.heatmap(similarity_df, annot=True, cmap='coolwarm')

plt.title("User Similarity Heatmap")
plt.xlabel("Users")
plt.ylabel("Users")

plt.show()

# =========================================
# 5. Recommendation Example
# =========================================
user = "User1"

similar_users = similarity_df[user].sort_values(ascending=False)

print("\nUsers similar to", user, ":\n")
print(similar_users)

# =========================================
# 6. Bar Graph of Similarity
# =========================================
plt.figure()

similar_users.plot(kind='bar')

plt.title(f"Similarity Scores for {user}")
plt.xlabel("Users")
plt.ylabel("Similarity Score")

plt.show()

# =========================================
# DONE
# =========================================
print("\n✅ Lab 12 executed successfully with visualization")
```
